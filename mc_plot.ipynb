```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("run_3.csv")
data.head()
```

```python
fig, ax = plt.subplots()
ax.plot(data.radius, data.variable_epsilon_regular_porosity, label="Square Lattice Packing")
ax.plot(data.radius, data.variable_epsilon_offset_porosity, label="Hexagonal Lattice Packing")
ax.plot(data.radius, data.variable_epsilon_lowest_first_porosity, label="Lowest First Packing")
ax.plot(data.radius, data.variable_epsilon_closest_first_porosity, label="Closest First Packing")
ax.set_xlabel("Radius [M]")
ax.set_ylabel("Porosity")
ax.set_ylim((0, 0.5))
ax.set_xlim((data.radius.min(), data.radius.max()))
plt.legend()
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(data.radius, data.variable_epsilon_regular_conductivity, label="Square Lattice Packing")
ax.plot(data.radius, data.variable_epsilon_offset_conductivity, label="Hexagonal Lattice Packing")
ax.plot(data.radius, data.variable_epsilon_lowest_first_conductivity, label="Lowest First Packing")
ax.plot(data.radius, data.variable_epsilon_closest_first_conductivity, label="Closest First Packing")
ax.set_xlabel("Radius [M]")
ax.set_ylabel("Thermal Conductivity [W/M]")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(data.variable_epsilon_regular_porosity, data.variable_epsilon_regular_conductivity, ".", label="Square Lattice Packing")
ax.plot(data.variable_epsilon_offset_porosity, data.variable_epsilon_offset_conductivity, ".", label="Hexagonal Lattice Packing")
ax.plot(data.variable_epsilon_lowest_first_porosity, data.variable_epsilon_lowest_first_conductivity, ".", label="Lowest First Packing")
ax.plot(data.variable_epsilon_closest_first_porosity, data.variable_epsilon_closest_first_conductivity, ".", label="Closest First Packing")
ax.set_xlabel("Porosity")
ax.set_ylabel("Thermal Conductivity [W/M]")
plt.legend()
plt.tight_layout()
plt.show()
```
