```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
radius = 5

fig, ax = plt.subplots()

for i in np.arange(0, 100, radius*2):
    for j in np.arange(0, 100, radius*2):
        c = patches.Circle((i+radius,j+radius), radius)
        ax.add_patch(c)

ax.set_xlim(0,100)
ax.set_ylim(0,100)
ax.set_aspect('equal')
```

```python
fig, ax = plt.subplots()

for i,y in enumerate(np.arange(0, 100-radius, radius*2*np.sin(np.pi/3))):
    if i % 2 == 0:
        lim = (0,100)
    else:
        lim = (radius,100-radius)
    for x in np.arange(*lim, radius*2):
        c = patches.Circle((x+radius,y+radius), radius)
        ax.add_patch(c)

ax.set_xlim(0,100)
ax.set_ylim(0,100)
ax.set_aspect('equal')
```

```python
r_minor = 5
r_major = 10
fig, ax = plt.subplots()

for x in np.arange(0, 100, r_minor*2):
    for y in np.arange(0, 100, r_major*2):
        c = patches.Ellipse((x+r_minor,y+r_major), 2*r_minor, 2*r_major)
        ax.add_patch(c)

ax.set_xlim(0,100)
ax.set_ylim(0,100)
ax.set_aspect('equal')
```
