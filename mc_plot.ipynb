```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("run_2.csv")
data.head()
```

```python
fig, ax = plt.subplots()

ax.plot(data.variable_epsilon_regular_porosity, 1/data.variable_epsilon_regular_resistance, "*", label="Regular")
ax.plot(data.variable_epsilon_offset_porosity,1/ data.variable_epsilon_offset_resistance, "*", label="Offset")
ax.plot(data.variable_epsilon_lowest_first_porosity, 1/data.variable_epsilon_lowest_first_resistance, "*", label="Lowest First")
ax.plot(data.variable_epsilon_closest_first_porosity, 1/data.variable_epsilon_closest_first_resistance, "*", label="Closest First")

ax.legend()
ax.set_xlabel("Porosity")
ax.set_ylabel("Thermal Conductance")
plt.show()
```
