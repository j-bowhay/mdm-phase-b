```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
radius = 5

fig, ax = plt.subplots()

for i in np.arange(0, 100, radius*2):
    for j in np.arange(0, 100, radius*2):
        c = patches.Circle((i+radius,j+radius), radius)
        ax.add_patch(c)

ax.set_xlim(0,100)
ax.set_ylim(0,100)
ax.set_aspect('equal')
```

```python
fig, ax = plt.subplots()

for i,y in enumerate(np.arange(0, 100-radius, radius*2*np.sin(np.pi/3))):
    if i % 2 == 0:
        lim = (0,100)
    else:
        lim = (radius,100-radius)
    for x in np.arange(*lim, radius*2):
        c = patches.Circle((x+radius,y+radius), radius)
        ax.add_patch(c)

ax.set_xlim(0,100)
ax.set_ylim(0,100)
ax.set_aspect('equal')
```

```python
r_minor = 5
r_major = 10
fig, ax = plt.subplots()

for x in np.arange(0, 100, r_minor*2):
    for y in np.arange(0, 100, r_major*2):
        c = patches.Ellipse((x+r_minor,y+r_major), 2*r_minor, 2*r_major)
        ax.add_patch(c)

ax.set_xlim(0,100)
ax.set_ylim(0,100)
ax.set_aspect('equal')
```

```python
r_minor = 5
r_major = 10
fig, ax = plt.subplots()

for i,y in enumerate(np.arange(0, 100, r_major)):
    if i % 2 == 0:
        lims = (0, 100, r_minor*2)
    else:
        lims = (0, 100, r_major*2)
    for x in np.arange(*lims):
        if i % 2 == 0:
            c = patches.Ellipse((x+r_minor,y+r_major), 2*r_minor, 2*r_major)
        else:
            c = patches.Ellipse((x+r_major,y+r_minor), 2*r_major, 2*r_minor)
        ax.add_patch(c)

ax.set_xlim(0,100)
ax.set_ylim(0,100)
ax.set_aspect('equal')
```

```python
import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection
from scipy.stats import qmc
rng = np.random.default_rng()
radius = 0.1
engine = qmc.PoissonDisk(d=2, radius=radius, ncandidates=1000, seed=rng)
sample = engine.fill_space()
fig, ax = plt.subplots()


_ = ax.scatter(sample[:, 0], sample[:, 1])
circles = [plt.Circle((xi, yi), radius=radius/2, fill=False)
           for xi, yi in sample]
collection = PatchCollection(circles, match_original=True)
ax.add_collection(collection)
_ = ax.set(aspect='equal', xlabel=r'$x_1$', ylabel=r'$x_2$',
           xlim=[0, 1], ylim=[0, 1])
plt.show()
```
