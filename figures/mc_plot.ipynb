```python
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science','ieee'])

data = pd.read_csv("run_5.csv")
data.head()
```

```python
fig, ax = plt.subplots()
ax.plot(data.radius, data.variable_epsilon_regular_porosity, label="Square Lattice")
ax.plot(data.radius, data.variable_epsilon_offset_porosity, label="Hexagonal Lattice")
ax.plot(data.radius, data.variable_epsilon_lowest_first_porosity, label="Lowest First")
ax.plot(data.radius, data.variable_epsilon_closest_first_porosity, label="Closest First")
ax.set_xlabel("Radius")
ax.set_ylabel("Porosity")
ax.set_xlim((data.radius.min(), data.radius.max()))
ax.set_ylim((0.08,0.65))
plt.legend()
plt.savefig("porosity_radius.pdf")
```

```python
fig, ax = plt.subplots()
ax.plot(data.radius, data.variable_epsilon_regular_conductivity, label="Square Lattice")
ax.plot(data.radius, data.variable_epsilon_offset_conductivity, label="Hexagonal Lattice")
ax.plot(data.radius, data.variable_epsilon_lowest_first_conductivity, label="Lowest First")
ax.plot(data.radius, data.variable_epsilon_closest_first_conductivity, label="Closest First")
ax.set_xlabel("Radius")
ax.set_ylabel("Thermal Conductance [W/M]")
ax.set_xlim((data.radius.min(), data.radius.max()))
ax.set_ylim((0,45))
plt.legend()
plt.savefig("radius_conductivity.pdf")
```

```python
fig, ax = plt.subplots()
ax.plot(data.variable_epsilon_regular_porosity, data.variable_epsilon_regular_conductivity, ".", label="Square Lattice")
ax.plot(data.variable_epsilon_offset_porosity, data.variable_epsilon_offset_conductivity, ".", label="Hexagonal Lattice")
ax.plot(data.variable_epsilon_lowest_first_porosity, data.variable_epsilon_lowest_first_conductivity, ".", label="Lowest First")
ax.plot(data.variable_epsilon_closest_first_porosity, data.variable_epsilon_closest_first_conductivity, ".", label="Closest First")
ax.set_xlabel("Porosity")
ax.set_ylabel("Thermal Conductance [W/M]")
ax.set_ylim([0,50])
plt.legend()
plt.savefig("porosity_conductivity.pdf")
```
