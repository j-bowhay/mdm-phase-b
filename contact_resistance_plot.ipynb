```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science','ieee'])

fig, axs = plt.subplots()
r = np.linspace(0,0.2, num=500)
resistance = (1/(4*np.pi*r)) * np.log((r*(r/100 - r))/(r/100 * (r - 2*r)))
axs.plot(r, resistance)
axs.set_xlim((0,0.2))
axs.set_xlabel("$r$")
axs.set_ylabel("Contact Resistance [K/W]")
plt.tight_layout()
plt.savefig("r_r.pdf")
```

```python
fig, axs = plt.subplots()
r = 0.1
epsilon = np.linspace(0,r/2, num=1000)
resistance = (1/(4*np.pi*r)) * np.log((r*(epsilon - r))/(epsilon * (r - 2*r)))
axs.plot(epsilon, resistance)
axs.set_xlim((0,r/2))
axs.set_ylim((0,6))
axs.set_xlabel("$\epsilon$")
axs.set_ylabel("Contact Resistance [K/W]")
plt.tight_layout()
plt.savefig("r_e.pdf")
```
